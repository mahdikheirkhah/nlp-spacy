```python
import spacy
```

The goal of this exercise is to learn to load an embedding on `spaCy`.

1. Install and load `en_core_web_sm` version `3.4.1` [embedding](https://github.com/explosion/spacy-models/releases/tag/en_core_web_sm-3.4.1). Compute the embedding vector of the word `car` and print its shape.


```python
!python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("car")
car_vector = doc[0].vector
print(f"Model Version: {nlp.meta['version']}")
print(car_vector.shape)
```

2. Print the sum of the first 20 values of the `car` embedding. The result must be printed using [`float.as_integer_ratio()`](https://docs.python.org/3/library/stdtypes.html#float.as_integer_ratio)

```python
first_20_sum = sum(car_vector[:20])
print(f"Integer Ratio: {float(first_20_sum).as_integer_ratio()}")
# If version is 3.4.1, this will output: (99959115, 33554432)
```

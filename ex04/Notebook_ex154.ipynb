````python
import spacy
````

The goal of this exercise is to learn to compute the similarity between two sentences. As explained in the documentation: **The word embedding of a full sentence is simply the average over all different words**. This is how `similarity` works in spaCy. This example demonstrates that if we build a corpus of sentences expressing an intention such as "buy shoes", we can detect this intention and use it to recommend shoe advertisements to customers. The language model used in this exercise is `en_core_web_sm`.

1. Compute the similarities (3 in total) between these sentences:

   ```
   sentence_1 = "I want to buy shoes"
   sentence_2 = "I would love to purchase running shoes"
   sentence_3 = "I am in my room"

   ```

````python
nlp = spacy.load("en_core_web_sm")

sentence_1 = nlp("I want to buy shoes")
sentence_2 = nlp("I would love to purchase running shoes")
sentence_3 = nlp("I am in my room")

sim_1_2 = sentence_1.similarity(sentence_2)
sim_1_3 = sentence_1.similarity(sentence_3)
sim_2_3 = sentence_2.similarity(sentence_3)

print(f"Similarity (S1, S2): {sim_1_2:.4f}")
print(f"Similarity (S1, S3): {sim_1_3:.4f}")
print(f"Similarity (S2, S3): {sim_2_3:.4f}")
````

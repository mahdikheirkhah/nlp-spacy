````python
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
````

The goal of this exercise is to learn to tokenize a document using `spaCy`. Previously, we did this using NLTK.

1. Tokenize the text below into individual tokens and print them.

   ```
       text = "Tokenize this sentence. And this one too."

   ```


````python
text = "Tokenize this sentence. And this one too."
nlp = English()
tokenizer = Tokenizer(nlp.vocab)
print(tokenizer.pipe(text))
````

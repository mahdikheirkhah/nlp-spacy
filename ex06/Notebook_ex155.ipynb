The goal of this exercise is to learn to use the Part-of-speech (POS) tags using `spaCy`. As explained on Wikipedia, POS tagging is the process of marking up a word in a text (corpus) as corresponding to a particular part of speech, based on both its definition and its context.

**Example of POS tagging ambiguity:**

The sentence: **"Heat water in a large vessel"**

Many words in English can serve as different parts of speech depending on context. Here's how spaCy tags this sentence:

| Word   | POS Tag in This Sentence | Alternative POS (in other contexts)                |
| ------ | ------------------------ | -------------------------------------------------- |
| heat   | VERB                     | NOUN (e.g., "The heat is unbearable")              |
| water  | NOUN                     | VERB (e.g., "Water the plants")                    |
| in     | ADP (preposition)        | NOUN (e.g., "the in-crowd"), ADV (e.g., "come in") |
| a      | DET (determiner)         | —                                                  |
| large  | ADJ (adjective)          | —                                                  |
| vessel | NOUN                     | —                                                  |

**Key insight:** The same word (e.g., "heat", "water") can have different grammatical roles depending on how it's used in the sentence. spaCy determines the correct tag based on context.

The file [`news_amazon.txt`](./resources/news_amazon.txt) contains newspaper articles about Amazon.

1. Return all sentences mentioning **Bezos** as a proper noun (POS tag: `PROPN` or detailed tag: `NNP`).

**Hint:** Use `token.pos_` for coarse-grained tags (`PROPN`, `NOUN`, `VERB`) or `token.tag_` for fine-grained Penn Treebank tags (`NNP`, `NN`, `VBZ`).

```python
import spacy

nlp = spacy.load("en_core_web_md")

with open("./resources/news_amazon.txt", "r") as file:
    text = file.read()

doc = nlp(text)

bezos_sentences = []

for sent in doc.sents:
    found_bezos = any(token.text == "Bezos" and token.pos_ == "PROPN" for token in sent)
    
    if found_bezos:
        bezos_sentences.append(sent.text.strip())

print(f"Found {len(bezos_sentences)} sentences mentioning Bezos as a PROPN:\n")
for i, s in enumerate(bezos_sentences, 1):
    print(f"{i}. {s}\n")
```
